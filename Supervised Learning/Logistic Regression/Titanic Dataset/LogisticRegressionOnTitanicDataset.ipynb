```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
pd.set_option('display.max_rows', None)
```

```python
training_data = pd.read_csv('training_titanic_x_y_train.csv')
```

```python
training_data.info()
```

```python
training_data.drop(['Cabin', 'Ticket', 'Name'], inplace = True, axis = 1)
```

```python
training_data.at[578, 'Embarked'] = training_data['Embarked'].mode()[0]
```

```python
training_data.fillna(value = {'Age' : training_data['Age'].mean()}, inplace = True)
```

```python
training_data.info()
```

```python
training_data
```

```python
X = training_data.iloc[:, :-1].values 
Y = training_data.iloc[:, -1].values 
```

```python
onehotencoder = OneHotEncoder()
X = np.append(X, onehotencoder.fit_transform(X[:, 0].reshape(-1,1)).toarray(), axis = 1)
X = np.append(X, onehotencoder.fit_transform(X[:, 1].reshape(-1,1)).toarray(), axis = 1)
X = np.append(X, onehotencoder.fit_transform(X[:, 6].reshape(-1,1)).toarray(), axis = 1)
```

```python
X = np.delete(X, [0,1,6], axis = 1)
```

```python
temp = []
for i in range(len(X[0])):
    for j in range(i, len(X[0])):
        temp.append(X[:, i]*X[:, j])
temp = np.array(temp)
for i in temp:
    X = np.append(X, i.reshape(-1, 1), axis = 1)
```

```python
scaler = StandardScaler() 
```

```python
scaler.fit(X)
X = scaler.transform(X)
```

```python
classifier = LogisticRegression(max_iter = 10000, C = 0.1)
```

```python
classifier.fit(X, Y)
```

```python
classifier.score(X, Y)
```
