```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def step_gradient(data, learning_rate, m, c):
    X = data[:, 0]
    Y = data[:, 1]
    
    M = len(data)
    
    slope_m = (-2/M)*(((Y - m * X - c) * X).sum())
    slope_c = (-2/M)*((Y - m * X - c).sum())
    
    new_m = m - learning_rate * slope_m
    new_c = c - learning_rate * slope_c
    
    return new_m, new_c
```

```python
def cost(data, m, c):
    X = data[:, 0]
    Y = data[:, 1]
    
    M = len(data)
    
    cost = (((Y - m * X - c)**2).sum())/M
    return cost
```

```python
def gradient_descent(data, learning_rate, num_iterations):
    m = 0
    c = 0
    total_cost = []
    for i in range(num_iterations):
        m, c = step_gradient(data, learning_rate, m, c)
        total_cost.append(cost(data, m, c))
    plt.plot(np.array(total_cost))
    plt.show()
    return m, c
```

```python
def run():
    data = np.loadtxt('data.csv', delimiter = ',')
    learning_rate = 0.0001
    num_iterations = 10000
    m, c = gradient_descent(data, learning_rate, num_iterations)
    print('m:', m, 'c:', c)
    X = data[:, 0]
    Y = data[:, 1]
    
    Y_predicted = (m * X + c)
    
    #COD
    u = ((Y - Y_predicted)**2).sum()
    v = ((Y - Y.mean())**2).sum()
    cd = 1 - (u/v)
    print('Coefficient of Determination:', cd)
    
    #plot
    plt.plot(Y, label = 'Actual')
    plt.plot(Y_predicted, label = 'Predicted')
    plt.legend()
    plt.show()
```

```python
run()
```
